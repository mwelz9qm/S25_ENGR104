So now we'll read in the states data file as a data frame and explore it. The states data file is txt with white space between columns -- we'll import that a little differently than a csv.

We'll read in the sensor data from last class, but in excel form instead.

````python
df = pd.read_excel('sensor_data.xlsx', usecols = 'A:F')
df
````

use .iloc to refer to specific row and index numbers

````python
df.iloc[3,3]  #entry in row 3, col 3
df.iloc[3,2:5]  #row 3, cols 2 to 5 (non-inclusive of 5)
````

If we want, we can use .loc to refer to data by labels instead of row/col numbers.
.loc can be compined with logic, where & means "and" and | means 'or'

````python
sd.loc[sd.Year > 1850]  #states founded after 1850
````

````python
sd.loc[(sd.Year > 1850) & (sd.Population < 3000000)] #states founded after 1850 and with population less than 3,000,000
````

````python
df.loc[(df.Sensor_ID == 'Sensor_1') | (df.Sensor_ID == 'Sensor_2'),'Sensor_ID':'PM2.5'] #filters for sensor1 or 2 and returns columns from sensor id to pm2.5
````
